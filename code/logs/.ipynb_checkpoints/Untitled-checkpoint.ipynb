```python
from matplotlib import pyplot as plt


def draw(file_names):
    f_name = [f.split(".")[0] for f in file_names]
    
    for i, file_name in enumerate(file_names):
        f = open(file_name)
        lines = f.readlines()
        f1 = []
        auc= []
        ignore_f1 = []
        ignore_AUC = []
        for line in lines:
            if "ALL  " in line: #Ignore 
                try:
                    pos_e = line.find('F1')
                    if i == len(file_names)-1:
                        f1.append(float(line[pos_e+3:pos_e+9]))
                        auc.append(float(line[pos_e+16:pos_e+22]))
                        
                    else:
                        f1.append(float(line[pos_e+3:pos_e+9]))
                        auc.append(float(line[pos_e+16:pos_e+22]))

                except:
                    break
        print(f1)
        plt.plot(f1, label=f_name[i])
#         plt.plot(auc[20:], label=file_name+'auc')
    plt.legend(bbox_to_anchor=(1.05, 1), fontsize=20)
    plt.show()
    
draw(["train_GCN4_SBloss_0_3_SBsize_3.log", "train_GCN6_SBloss_0_3_SBsize_3.log"])
```
